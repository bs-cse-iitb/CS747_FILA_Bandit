```python
def kl(x,y):
    if x==0:
        kl_value = -math.log((1-y))
        return kl_value
    if x==1:
        kl_value = -math.log(y)
        return kl_value
    else:
        ln1 = math.log(x/y)
        ln2 = math.log((1-x)/(1-y))
        kl_value = x * ln1 + (1-x) * ln2 
        return kl_value

def bs(p_,start,end,kl_p_q):
    middle = (start + end)/2
    #print(middle)
    kl_p_m = kl(p_,middle)
    print(kl_p_m , kl_p_q)
    #return 0.5
    if abs(kl_p_m-kl_p_q) < 0.0001:
        print(middle)
        return middle
    elif(kl_p_m > kl_p_q ):
        bs(p_,start,middle,kl_p_q)
    else:
        bs(p_,middle,end,kl_p_q)

```

```python
class Algorithm:
    def __init__(self, num_arms, horizon):
        self.num_arms = num_arms
        self.horizon = horizon
    
    def give_pull(self):
        raise NotImplementedError
    
    def get_reward(self, arm_index, reward):
        raise NotImplementedError
```

```python
class UCB(Algorithm):
    def __init__(self, num_arms, horizon):
        super().__init__(num_arms, horizon)
        # You can add any other variables you need here
        # START EDITING HERE

        #num_arms no. of arms
        # ucb : ucb values of each arm
        self.ucb = np.zeros(num_arms)
        #counts : no. of pull of each arm
        self.counts = np.zeros(num_arms)
        # mean : empirical mean
        self.mean = np.zeros(num_arms)
        self.h = 0
        # END EDITING HERE
        # END EDITING HERE
    
    def give_pull(self):
        # START EDITING HERE
        #print(self.h)
        #print(self.num_arms)
        if (self.h < self.num_arms):
            return self.h
        return np.argmax(self.ucb)
        # END EDITING HERE
    
    def get_reward(self, arm_index, reward):
        # START EDITING HERE

        # empirical means of arm_index    
        p_ = self.mean[arm_index]
        u = self.counts[arm_index]
        self.mean[arm_index]= (p_ * u + reward ) / (u+1) 

        print("arm : {} rew : {} mean: {} ucb : {} ".format(arm_index,reward,self.mean, self.counts))

        self.counts[arm_index]+=1
        self.h+=1

        if (self.h > self.num_arms):
            print("check")
            ln = math.log(self.h)
            for arm in range(self.num_arms):
                add_to_mean = math.sqrt((2*ln)/self.counts[arm])
                self.ucb[arm]= self.mean[arm] + add_to_mean
        # END EDITING HERE
```

```python
horizon = 100
num_arms = 2
prob = [0.4,0.7]
np.random.seed(0)
```

```python
class KL_UCB(Algorithm):
    def __init__(self, num_arms, horizon):
        super().__init__(num_arms, horizon)
        # You can add any other variables you need here
        # START EDITING HERE
        #num_arms no. of arms

        self.kl_ucb = np.zeros(num_arms)
        #no. of pull of each arm
        self.counts = np.zeros(num_arms)
        self.mean = np.zeros(num_arms)
        self.h = 0
        # END EDITING HERE
    
    def give_pull(self):
        # START EDITING HERE
        # round robin two times 
        if (self.h<(self.num_arms+1)*2-1):
            return (self.h)%self.num_arms   
        return np.argmax(self.kl_ucb)
        # END EDITING HERE
    
    def get_reward(self, arm_index, reward):
        # START EDITING HERE
        # empirical means of arm_index    
        p_ = self.mean[arm_index]
        u = self.counts[arm_index]
        p_= (p_ * u + reward ) / (u+1)
        self.mean[arm_index]= p_  
        # END EDITING HERE

        self.counts[arm_index]+=1
        self.h+=1
        
        def bs(p_,start,end,kl_p_q):
            middle = (start + end)/2
            #print(middle)
            kl_p_m = kl(p_,middle)
            #print(kl_p_m , kl_p_q)
            #return 0.5
            if (abs(kl_p_m-kl_p_q)< 0.1):
                return middle
            elif(kl_p_m > kl_p_q):
                bs(p_,start,middle,kl_p_q)
            else:
                bs(p_,middle,end,kl_p_q)

        # mistake 1 >=  instead of >
        if (self.h > (self.num_arms+1)*2-1):

            ln_t = math.log(self.h)
            lnln_t = math.log(ln_t)
            c = 3

            q = 0.999
            for arm in range(self.num_arms):
                kl_p_q = (ln_t + c * lnln_t) / self.counts[arm]
                self.kl_ucb[arm] = bs(p_,p_,q,kl_p_q)
```

```python
ucb = UCB(num_arms,horizon)
for i in range(10):
    x  = ucb.give_pull()
    if np.random.uniform(0,1)<prob[x]:
        rew =1
    else:
        rew =0
        
    ucb.get_reward(x,rew)
    #print(x)
```

```python
class Thompson_Sampling(Algorithm):
    def __init__(self, num_arms, horizon):
        super().__init__(num_arms, horizon)
        # You can add any other variables you need here
        # START EDITING HERE
        #num_arms no. of arms
        #success no. of success of each arm
        self.success = np.zeros(num_arms)
        #no. of pull of each arm
        self.counts = np.zeros(num_arms)
        # END EDITING HERE
    
    def give_pull(self):
        # START EDITING HERE
        np.random.seed(0)
        #sample : array to store sample value from beta distribution
        sample = np.zeros(self.num_arms)
        #arm : index sarting from zero
        # for every arm a draw as sample from beta
        for arm in range(self.num_arms):
            sample[arm]=np.random.beta(self.success[arm]+1,self.counts[arm]-self.success[arm]+1)
        return np.argmax(sample)
        
        # END EDITING HERE
    
    def get_reward(self, arm_index, reward):
        # START EDITING HERE
        self.counts[arm_index]+=1
        if reward==1:
            self.success[arm_index]+=1
        print("arm : {} rew : {} success: {} count : {} ".format(arm_index,reward,self.success, self.counts))

        # END EDITING HERE
```
